### Transformação de Dados

```python
# Transformar coluna df_contracts["TotalCharges"] de String para Float - Abordagem 1 com astype
df_contracts.TotalCharges = df_contracts.TotalCharges.astype(float)
```

```python
# Transformar coluna df_contracts["TotalCharges"] de String para Float - Abordagem 2 com to_numeric
df_contracts.TotalCharges = pd.to_numeric(df_contracts.TotalCharges)
```

```python
# Transformar coluna df_contracts["TotalCharges"] de String para Float - Abordagem 3 com to_numeric com coerce
df_contracts.TotalCharges = pd.to_numeric(df_contracts.TotalCharges, errors='coerce')
```

### Unificar os 3 dataframes

```python
# Contar quantidade de registros por dataframe
len(df_customers)
```

```python
# Renomear coluna
df_services.rename(
    columns= {
        'customerID' : 'customer_id'
    },
    inplace=True
)
```

```python
# Unificar dataframe de customers com services, criando um terceiro dataframe
df_temp = df_customers.merge(df_services, on=['customer_id'])
```

```python
# mostrar estrutura df_tem
df_temp.info()
```

```python
# Unificar dataframe temp com contracts, criando um novo dataframe - usando colunas com nomes diferentes
df_churn_temp = df_temp.merge(df_contracts, left_on=['customer_id'], right_on=['customerID'])
```

```python
# Unificar os tres dataframes ao mesmo tempo, com colunas com nomes diferentes
df_churn = df_customers.merge(df_services, on=['customer_id']).merge(df_contracts, left_on=['customer_id'], right_on=['customerID'])
```

```python
df_churn.info()
```

```python
# Remover coluna de um dataframe
df_churn.drop(columns=['customerID'], axis=1, inplace=True)
```

```python
# Histograma do Tempo de contrato
df_churn.tenure.plot.hist()
```

- A váriavel tempo de contrato (tenure), não apresenta visualmente uma distribuição do tipo normal.
- Grande parte dos contratos se concentram nos extremos, ou seja, contratos novos ou mais antigos com mais de 65 meses.

```python
# Histograma do Monthly charges
df_churn.MonthlyCharges.plot.hist()
```

- Maioria dos contratos tem valores mais baixo, isso pode estar relacionado com a maioria dos contratos sendo de serviços mais básico
- Muito contratos com valores entre 75 e 100 dolares

```python
# Medidas de Posiçõa - Média de Tempo de Contrato
df_churn.tenure.mean()
```

```python
# Medidas de Posiçõa - Mediana de Tempo de Contrato
df_churn.tenure.median()
```

```python
# Medidas de Posiçõa - Moda de Tempo de Contrato
df_churn.tenure.mode()
```

```python
# Medidas de dispersão - Desvio padrão Tempo de Contrato
df_churn.tenure.std()
```

```python
# Medidas de dispersão - Coeficiente de Variação Tempo de Contrato
df_churn.tenure.std() / df_churn.tenure.mean() * 100
```

- Moda = 1 - muitos contratos novos
- Em média, os contratos estão por volta de 30 meses, porém o desvio padrão é muito alto indicando que esse média tem forte influência de outliers
- Analisando o histograma, temos uma grande concentração de de contratos com mais de 65 meses de duração, mas a grande concentração de contratos com 1 mês de duração move a mediana e a média para mais perto de 1 mes
